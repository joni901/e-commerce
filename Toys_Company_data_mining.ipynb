```python
with open(r'/content/sales_data_sample (1).csv', "r",encoding='latin1') as file:
  r_file = file.readlines()
  files=[]
  for line in r_file:
    substrings = line.split(',')
    files.append(substrings)
  print(files)

```

```python
table = {files[0][i]: [row[i] for row in files[2:]] for i in range(len(files[0]))}
table
```

```python
import pandas as pd

df=pd.DataFrame(table)
```

```python
df.shape
```

```python
df.info()
```

```python
count=(df=='').sum().sum()
count
```

```python
for d in df.columns:
  count_d=(df[d]=='').sum()
  print(f'{d}:{count_d}')
```

```python
group_result_postalcode = df.groupby('POSTALCODE')['CITY'].apply(lambda x: (x == "").sum()).reset_index(name='zero_city')

for index, row in group_result_postalcode.iterrows():
    if row['zero_city'] != 0:
        print(f"Postal Code {row['POSTALCODE']} has {row['zero_city']} empty CITY values.")
```

```python
group_result_city = df.groupby('CITY')['POSTALCODE'].apply(lambda x: (x == "").sum()).reset_index(name='zero_code')

for index, row in group_result_city.iterrows():
    if row['zero_code'] != 0:
        print(f"City {row['CITY']} has {row['zero_code']} empty postal code values.")
```

```python
group_result_address = df.groupby('ADDRESSLINE1')['POSTALCODE'].apply(lambda x: (x == "").sum()).reset_index(name='zero_address')

for index, row in group_result_address.iterrows():
    if row['zero_address'] != 0:
        print(f"Address {row['ADDRESSLINE1']} has {row['zero_address']} empty postal code values.")
```

```python
df.columns = df.columns.str.strip()

# Agora, faça o agrupamento
contagem_pares = df.groupby([ 'PRODUCTLINE','ORDERLINENUMBER']).size().reset_index(name='Contagem')

# Imprima o resultado
pd.set_option('display.max_rows', None)
print(contagem_pares)
pd.set_option('display.max_rows', None)
```
